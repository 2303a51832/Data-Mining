```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df=pd.read_csv('/content/dataset.csv',delimiter=';')
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,15))
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='quality', y='fixed acidity', data=df, marker='o', ci=None)
plt.title('Line Plot: Fixed Acidity vs. Quality', fontsize=16)
plt.xlabel('Quality', fontsize=14)
plt.ylabel('Fixed Acidity', fontsize=14)
plt.grid(True)  # Add gridlines for better readability
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='quality', y='alcohol', data=df, hue='quality', palette='cool', s=100)
plt.title('Scatter Plot: Alcohol vs. Quality', fontsize=16)
plt.xlabel('Quality', fontsize=14)
plt.ylabel('Alcohol', fontsize=14)
plt.legend(title='Quality', fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='quality', y='sulphates', data=df, palette='viridis')
plt.title('Bar Plot: Average Sulphates by Quality')
plt.xlabel('Quality')
plt.ylabel('Sulphates')
```

```python
# 4. Histogram: Distribution of 'alcohol'
plt.figure(figsize=(10, 6))
sns.histplot(df['alcohol'], kde=True, bins=20, color='teal')
plt.title('Histogram: Distribution of Alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10, 6))
quality_counts = df['quality'].value_counts()
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Pie Chart: Proportion of Quality Scores')
```
